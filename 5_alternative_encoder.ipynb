```python
def encoder_5q(x1, x2):
    qc = QuantumCircuit(5)
    if x1 == 0 and x2 == 1:
        qc.x(4)
    elif x1 == 1 and x2 == 0:
        qc.h(4)
    elif x1 == 1 and x2 == 1:
        qc.x(4)
        qc.h(4)

    qc.x(0)
    qc.x(3)
    qc.barrier()
    qc.cx(4, 0)
    qc.h(0)
    qc.h(1)
    qc.cx(1, 4)
    qc.cx(1, 2)
    qc.cx(0, 2)
    qc.cx(4, 3)
    qc.cx(0, 4)
    qc.barrier()
    qc.h(2)
    qc.h(3)
    qc.cx(3, 4)
    qc.swap(2, 1)
    qc.cx(1, 4)

    return qc
```
